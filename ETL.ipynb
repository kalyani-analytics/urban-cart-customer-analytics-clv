```python
#Extract Data from mongo db
from pymongo import MongoClient
import pandas as pd

# MongoDB connection
client = MongoClient("mongodb://localhost:27017/")

db = client["UrbanCart_logs"]
collection = db["web_logs"]

# Extract documents
mongo_data = list(collection.find({}, {"_id": 0}))
web_logs_df = pd.DataFrame(mongo_data)

print(web_logs_df.head())
```

```python
#count How many duplicate row exit in customer table
customers_df.duplicated().sum()
```

```python
# Convert signup_date to datetime
customers_df['signup_date'] = pd.to_datetime(customers_df['signup_date'])

# Standardize text columns
customers_df['gender'] = customers_df['gender'].str.strip().str.title()
customers_df['city'] = customers_df['city'].str.strip().str.title()
```

```python
#count How many duplicate row exit in orders table
orders_df.duplicated().sum()
```

```python

# Convert order_date to datetime
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])

# Remove invalid values
orders_df = orders_df[
    (orders_df['quantity'] > 0) &
    (orders_df['price'] > 0) &
    (orders_df['amount'] > 0)
]

# Standardize channel names
orders_df['channel'] = orders_df['channel'].str.strip().str.title()
```

```python
#count How many duplicate row exit in products table
products_df.duplicated().sum()
```

```python
# Standardize category
products_df['category'] = products_df['category'].str.strip().str.title()

# Remove invalid product pricing
products_df = products_df[products_df['price'] >= products_df['cost']]
```

```python
#count How many duplicate row exit in marketing_campaigns table
marketing_campaigns_df.duplicated().sum()
```

```python
# Convert period to datetime
marketing_campaigns_df['period'] = pd.to_datetime(marketing_campaigns_df['period'])

# Remove invalid spend
marketing_campaigns_df = marketing_campaigns_df[marketing_campaigns_df['spend'] >= 0]

# Standardize channel
marketing_campaigns_df['channel'] = marketing_campaigns_df['channel'].str.strip().str.title()
```

```python
#count How many duplicate row exit in web_logs table
web_logs_df.duplicated().sum()
```

```python
# Convert session_date to datetime
web_logs_df['session_date'] = pd.to_datetime(web_logs_df['session_date'])

# Remove invalid engagement values
web_logs_df = web_logs_df[
    (web_logs_df['pages_viewed'] >= 0) &
    (web_logs_df['time_spent_seconds'] >= 0)
]
```

```python
print("CUSTOMERS")
print(customers_df.info())

print("\nORDERS ")
print(orders_df.info())

print("\nPRODUCTS ")
print(products_df.info())

print("\nMARKETING_CAMPAIGNS ")
print(marketing_campaigns_df.info())

print("\nWEB_LOGS")
print(web_logs_df.info())
```

```python
#converts raw web session data into customer-level engagement data.
#Each customer may have many sessions, but after this code:
#1 row = 1 customer
web_engagement = web_logs_df.groupby('customer_id').agg(
    avg_pages_viewed=('pages_viewed', 'mean'),
    avg_time_spent=('time_spent_seconds', 'mean')
).reset_index()
```
